# Data Cleaning (Exploratory Data Analysis)

```python
import pandas as pd

df1 = pd.read_csv('data.csv')
```

```python
print(df1.to_string())
```

```python
df = df1.copy()
```

```python
### Info about the data
print(df.info()) # also indicates null values
```

```python
# Null Values
print(df.isnull().sum())
```

### Dealing with empty cells

1. **Remove rows**

```python

new_df = df.dropna()

print(new_df.to_string())
```

```python
print(new_df.isnull().sum())
```

```python
print(new_df.info())
```

   **Remove and add inplace.**

```python
df = pd.read_csv('data.csv')
df.dropna(inplace = True)

print(df.to_string())
```

2. **Replace empty values.**

```python
df = pd.read_csv('data.csv')
df.fillna(130, inplace = True)
```

```python
print(df.to_string())
```

2. **Replace only specific values**

```python
df = pd.read_csv('data.csv')
df["Calories"].fillna(130, inplace = True)
```

3. **Replace with mean, median, mode**

```python
df = pd.read_csv('data.csv')

#calculate mean
x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)
```

```python
print(df.to_string())
```

```python
df = pd.read_csv('data.csv')

#Median
x = df["Calories"].median()

df["Calories"].fillna(x, inplace = True)
```

```python
print(df.to_string())
```

```python
df = pd.read_csv('data.csv')

# Mode
x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)
```

```python
print(df.to_string())
```
