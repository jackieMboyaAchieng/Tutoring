```python
# import important libraries
import pandas

# create a dictionary

car_purchase = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2],
    'color': ['red', 'blue', 'green']
}
print(car_purchase)
```

```python
#printing pandas dataframe
car_purchase1 = pandas.DataFrame(car_purchase)
print(car_purchase1)
```

```python
car_purchase1.loc[0]
```

```python
car_purchase1.iloc[0, 1]
```

```python
car_purchase1.iloc[[0, 2], 0]
```

```python
#Naming indexes
import pandas as pd
car_purchase = pd.DataFrame(car_purchase, index = ["a", "b", "c"])
print(car_purchase)
```
